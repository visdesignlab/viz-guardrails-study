```python
import pandas as pd
```

```python
df = pd.read_csv('./owid-covid-data.csv')
df
```

```python
dict = pd.read_csv('./dict.csv')
dict
```

```python
num_dict = dict.to_dict()['location']
num_dict = {v: k+1 for k, v in num_dict.items()}
num_dict
```

```python
df_clean = df[['continent', 'location', 'date', 'new_deaths_smoothed_per_million']].copy()
df_clean = df_clean.rename(columns={'new_deaths_smoothed_per_million': 'value'})
df_clean = df_clean[df_clean['value']==df_clean['value']]
df_clean = df_clean[df_clean['location'].isin(dict['location'])]
df_clean['country_num'] = df_clean['location'].map(lambda x: num_dict[x])

df_clean = df_clean.merge(dict, on='location', how='left')
df_clean = df_clean.sort_values(by = ['country_num', 'date'])
df_clean = df_clean.reset_index(drop=True)

df_clean['name'] = df_clean['country_name'] + ' (Policy ' + df_clean['policy'] + ')'

df_clean
```

```python
df_clean.to_csv('./clean_data.csv')
```
